```python
from geopy.geocoders import Nominatim

def geocoding_reverse_state(lat, long):
    # geocoding으로 위경도 데이터 넣어서 불러오기
    lat_lng_str = f'{lat}, {long}'
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    response = geolocoder.reverse(lat_lng_str)
    address_list = response.address.split(',')
    
    # 불러온 데이터를 전처리 : 공백없애기
    stripped_address_list = []
    for i in address_list:
        stripped_address_list.append(i.lstrip())
        
    # 리스트에서 도 찾아서 반환하기
    for state in stripped_address_list:
        if state.endswith('도'):
            return state
        elif state.endswith('광역시'):
            return state
        elif state.endswith('특별시'):
            return state
        
        
def geocoding_reverse_city(lat, long):
    lat_lng_str = f'{lat}, {long}'
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    response = geolocoder.reverse(lat_lng_str)
    address_list = response.address.split(',')
    stripped_address_list = []
    for i in address_list:
        stripped_address_list.append(i.lstrip())
        
    for city in stripped_address_list:
        if city.endswith('군'):
            return city
        elif city.endswith('구'):
            return city
        elif city.endswith('시'):
            return city
    

# responses = []

# for i, row in df.iterrows():
#     response = geocoding_reverse((row.lat, row.long))
#     responses.append(response)

# print(responses)
```
