### 안산시

```python
import pandas as pd
as_df = pd.DataFrame({'State':[], 'City':[], 'cnt':[], '기준일':[]})
```

```python
gg = pd.read_csv('./kyeonggi/gg_as.csv', encoding='cp949')
gg_as = gg[['데이터기준일자', '관리기관명']]
gg_date = gg_as['데이터기준일자'].head(1)
gg_state = gg_as['관리기관명'].str[:3].head(1)
gg_city = gg_as['관리기관명'].str[4:7].head(1)
```

```python
as_df['State'] = gg_state
as_df['City'] = gg_city
as_df['cnt'] = gg_as.shape[0]
as_df['기준일'] = gg_date
as_df
```
