```python
import requests

def coordinate_to_address(lat, lng):
    api_key = ''
    headers = {'Authorization': f'KakaoAK {api_key}'}
    url = f'https://dapi.kakao.com/v2/local/geo/coord2address.json?x={lng}&y={lat}&input_coord=WGS84'
    try:
        response = requests.get(url, headers=headers)
        address = response.json()['documents'][0]['address']['address_name']
        return address
    except:
        return None
    
def coordinate_to_gun(lat, lng):
    api_key = 'my_api_key'
    headers = {'Authorization': f'KakaoAK {api_key}'}
    url = f'https://dapi.kakao.com/v2/local/geo/coord2address.json?x={lng}&y={lat}&input_coord=WGS84'
    try:
        response = requests.get(url, headers=headers)
        gun = response.json()['documents'][0]['address']['region_2depth_name']
        return gun
    except:
        return None

```

```python
import googlemaps

def coordinate_to_address(lat,long) -> str:
    # 발급받은 API 키 입력
    api_key = 'my_api_key'

    # googlemaps 객체 생성
    gmaps = googlemaps.Client(api_key)

    # reverse_geocode() 함수를 이용하여 해당 위치의 주소를 불러옴
    results = gmaps.reverse_geocode((lat, long), language='ko')
    
    for result in results:
        try:
            test = result['address_components'][-3]['long_name']
            address = result['formatted_address']
            if (test.endswith('도')) or (test.endswith('광역시')) or (test.endswith('특별시')):
                return address
        except:
            return '뀨'
null_data_df['new_address'] = null_data_df.apply(lambda row: coordinate_to_address(row['lat'], row['long']), axis=1)

def coordinate_to_result(lat,long) -> str:
    # 발급받은 API 키 입력
    api_key = 'my_api_key'

    # googlemaps 객체 생성
    gmaps = googlemaps.Client(api_key)


    # reverse_geocode() 함수를 이용하여 해당 위치의 주소를 불러옴
    results = gmaps.reverse_geocode((lat, long), language='ko')
    return results

# 함수 적용해서 새로운 컬럼 생성
null_data_df['new_address'] = null_data_df.apply(lambda row: coordinate_to_address(row['lat'], row['long']), axis=1)
```
