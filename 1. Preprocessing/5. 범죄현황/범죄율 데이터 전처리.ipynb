```python
# 데이터 로드
df = pd.read_csv('범죄율.csv', encoding='cp949', skiprows=[0])  # skiprows=[0] : 첫 행 삭제
df
```

```python
# df_crime.columns
```

```python
# '형법범: 강력범죄, 재산번죄' 만 남기기
df2 = df[df['범죄별(1)'].str.contains('형법범')].reset_index()

crime_lists = ['재산범죄', '강력범죄(흉악)', '강력범죄(폭력)']
df2 = df2[df2['범죄별(2)'].isin(crime_lists)].reset_index()
df2.drop(['index', 'level_0'], axis=1, inplace=True)
# df2['범죄별(2)'].unique()
df2
```

```python
# str -> int
# lists = ['강원', '경기', '경남', '경북', '광주', '대구', '대전', '부산', '서울', '세종', '울산', '인천', '전남', '전북', '제주', '충남', '충북']

# # for l in lists:
# #     df2[l] = pd.to_numeric(df2[l], error='coerce')  # to_numeric을 이용하여 errors = coerce 옵션으로 변환할 수 없는 데이터는 NaN으로 바꾸기
# # #     df2[l] = df2[l].astype(int)
# for l in lists:
#     df2 = df2.astype({l : 'float'})

# df2
```

```python
df2.drop([0], inplace=True)  # 소계 행 삭제
# df2.drop(['범죄별(3)', '계'], axis=1)
df2
```

```python
df2.drop(['범죄별(3)', '계'], axis='columns', inplace=True)
# df2.drop('index')
df2 = df2.reset_index()
df2
```

```python
df2.drop(['index'], axis='columns', inplace=True)
df2
```

```python
## 데이터프레임 행열전환 (transpose)

# df_trans_before = df2[['범죄별(1)', '범죄별(2)']]
df_trans_before = df2[df2.columns.difference(['범죄별(1)', '범죄별(2)'])]
# len(df_trans_before.columns)
# df_trans_before = pd.to_numeric(df_trans_before)
df_trans_before
```

```python
df_trans_before.columns
```

```python
# 특수 문자 제거
df_trans_before["세종"]= df_trans_before["세종"].str.replace("-", '', 1) #첫 번째 하이픈(-) 제거 
df_trans_before["제주"]= df_trans_before["제주"].str.replace("-", '', 1) #첫 번째 하이픈(-) 제거 

df_trans_before
```

```python
# str -> int
from numpy import NaN, NAN, nan #누락값 넣기 위한 모듈 불러오기

df_trans_before = df_trans_before.replace(r'^\s*$',NaN, regex=True)
df_trans_before
```

```python
#누락값 처리 NaN을 다른 값으로 변경하기
df_trans_before = df_trans_before.fillna(0) #다른 값 변경시 .fillna(변경할 숫자).iloc[행위치, 열위치] 입력하여 변경
df_trans_before
```

```python
lists = ['강원', '경기', '경남', '경북', '광주', '대구', '대전', '부산', '서울', '세종', '울산', '인천', '전남', '전북', '제주', '충남', '충북']

for l in lists:
    df_trans_before[l] = df_trans_before[l].astype('int64')
df_trans_before

df_trans_before
```

```python
df_trans_before.info()
```

```python
# transpose
df_trans_after = df_trans_before.transpose()
df_trans_after
```

```python
# 범죄 합계 먼저 구하기
df_trans_after.loc[:, 'total_sum'] = df_trans_after.loc[:,'0':'14'].sum(axis=1)
df_trans_after.reset_index(inplace=True)
df_trans_after
```

```python
df_trans_after = df_trans_after.rename(columns={'index':'state'})
df_trans_after
```

```python
df_trans_after.columns
```

```python
df3 = df_trans_after[['state', 'total_sum']]
df3
```

```python
df3.dtypes
```

```python
# 총 범죄건수 변수 할당
total_crime = df3.total_sum.sum(axis=0)
total_crime
```

```python
# 지역별 범죄율 구하기
df3['crime_rate'] = round(df3['total_sum']/total_crime, 4)
df_crime = df3[['state', 'crime_rate']]
```

```python
# state 명 변경
'''
서울특별시
부산광역시
인천광역시
대구광역시
대전광역시
광주광역시
울산광역시
세종특별자치시
경기도
강원도
충청북도
전라북도
전라남도
경상북도
경상남도
제주특별자치도
'''
state_error = ['강원', '경기', '경남', '경북', '광주', '대구', '대전', '부산', '서울', '세종', '울산', '인천', '전남', '전북', '제주', '충남', '충북']
state_correct = ['강원도', '경기도', '경상남도', '경상북도', '광주광역시', '대구광역시', '대전광역시', '부산광역시', '서울특별시', '세종특별자치시', '울산광역시', '인천광역시', '전라남도', '전라북도', '제주특별자치도', '충청남도', '충청북도']

# 행정구역별 딕셔너리화
dict_state = dict(zip(state_error, state_correct))
dict_state
```

```python
df3 = df3.replace({'state': dict_state})
df3.drop(['crime_rate'], axis=1, inplace=True)
```
