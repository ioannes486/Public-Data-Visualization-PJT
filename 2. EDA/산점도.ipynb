```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

plt.rcParams['font.family'] = 'Malgun Gothic' # 윈도우즈의 '맑은 고딕'설정
plt.rcParams['axes.unicode_minus'] = False
```

```python
a = {
    'visit_cnt' : 'sum',
    'cctv_cnt' : 'sum',
    'sclight_cnt' : 'sum',
    'smalight_cnt' : 'sum',
    'mean_hours' : 'mean',
    'mean_days' : 'mean',
    'crime_rate' : 'mean',
    'avg_pop_density' : 'mean',
    
}


state_groupby = data_df.groupby(['state'], as_index=False).agg(a)
state_groupby
```

### 범죄율과 각 요소별 상관관계 - 산점도

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.rcParams['font.family'] = 'Malgun Gothic' # 윈도우즈의 '맑은 고딕'설정
plt.rcParams['axes.unicode_minus'] = False
fig,axes = plt.subplots(2,2,sharex=False, figsize=(16,16))

fig.suptitle('시도별 요소대비 범죄율 산점도')
axes[0][0].set_title('범죄율 vs cctv수')
axes[0][1].set_title('범죄율 vs 보안등수')
axes[1][0].set_title('범죄율 vs 비상벨수')
axes[1][1].set_title('범죄율 vs 방문자수')
sns.scatterplot(x='cctv_cnt', y='crime_rate',
           data=state_groupby,
            hue="state",
            palette='deep',
            ax=axes[0][0],
            legend='full',
            markers=True
          )

sns.scatterplot(x='sclight_cnt', y='crime_rate',
           data=state_groupby,
                hue="state",
            palette='deep',
            ax=axes[0][1],
            legend='full'
          )
sns.scatterplot(x='smalight_cnt', y='crime_rate',
           data=state_groupby,
                hue="state",
            palette='deep',
            ax=axes[1][0],
            legend='full'
          )
sns.scatterplot(x='visit_cnt', y='crime_rate',
           data=state_groupby,
                hue="state",
            palette='deep',
            ax=axes[1][1],
            legend='full'
          )
plt.subplots_adjust(top=0.95, bottom=0.1, left=0.1, right=0.9, wspace=0.1, hspace=0.2)

plt.savefig('scatterplot_crime_rate_vs_resources.png', dpi=300)
```

### matplotlib으로 산점도 그리기

```python
import matplotlib.pyplot as plt
import numpy as np
from adjustText import adjust_text
from random import randrange
crime_rate = state_groupby.crime_rate
cctv_cnt = state_groupby.cctv_cnt
sclight_cnt = state_groupby.sclight_cnt
smalight_cnt = state_groupby.smalight_cnt
strlamp_cnt = state_groupby.strlamp_cnt
avg_pop_density = state_groupby.avg_pop_density
states = state_groupby.state
colors = ['r','g','b','c','m','k','y']
c = [colors[randrange(len(colors))] for i in range(len(states))]


# 산점도 그리기
plt.scatter(cctv_cnt, crime_rate, c=c)
plt.xlabel('지역별 cctv개수')
plt.ylabel('지역별 범죄율')
plt.title('cctv개수 대비 범죄율 산포도')

# 텍스트 라벨 설정
kargs = {
    'size':'xx-small',
    'ha':'center',
    'va':'center_baseline',
}

texts = []
for x,y, state in zip(cctv_cnt, crime_rate, states):
    texts.append(plt.text(x,y, state, kargs))

# adjustText 패키지를 사용하여 텍스트 라벨 위치 조절
adjust_text(texts)

plt.show()
```
