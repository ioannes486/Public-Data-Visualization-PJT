```python
import os 
import glob
import pandas as pd
import numpy as np

folder_path = 'C:/Users/miinn/보안등'
files_path = 'C:/Users/miinn/보안등'
read_files = glob.glob(os.path.join(files_path, '*.csv'))
file_names = []
    
for files in read_files:
    file_names.append(files)
    
global file_names
```

```python
def f_sclight1(x):
    
    np_array_values = []
    security_light = pd.read_csv(file_names[x], encoding = 'cp949', header= 0)
    np_array_values.append(security_light)
    merge_values = np.vstack(np_array_values)
    df_sclight1 = pd.DataFrame(merge_values)
    df_sclight1.columns = security_light.columns
    df_sclight1_1 = df_sclight1.drop(['보안등위치명', '설치연도', '설치형태', '관리기관전화번호', '관리기관명'], axis = 1)
    df_sclight1_1['주소'] = np.where(pd.notnull(df_sclight1_1['소재지지번주소'])==True, df_sclight1_1['소재지지번주소'], df_sclight1['소재지도로명주소'])
    df_sclight1_2 = df_sclight1_1.drop(['소재지도로명주소', '소재지지번주소'], axis = 1)
    df_sclight1_2 = df_sclight1_2[['주소', '위도', '경도', '설치개수', '데이터기준일자']]
    df_sclight1_2.rename(columns={'주소': 'address', '위도': 'lat', '경도':'long', '설치개수':'install_cnt', '데이터기준일자':'date'}, inplace=True)
    df_sclight1_2['state'] = df_sclight1_2['address'].str.split().str[0]
    df_sclight1_2['city'] = df_sclight1_2['address'].str.split().str[1]
    df_sclight1_2 = df_sclight1_2[['address', 'state', 'city', 'lat', 'long', 'install_cnt', 'date']]
    
    print(f'last page: {len(file_names)-1}')
    return pd.DataFrame(df_sclight1_2)
```
