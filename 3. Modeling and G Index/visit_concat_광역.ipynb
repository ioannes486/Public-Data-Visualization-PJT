```python
import pandas as pd
visit_df = pd.read_csv('./20230304180709_광역별방문자수.csv', encoding='cp949')
visit_df
```

```python
visit_df.drop(columns=['광역지자체 방문자 비율', '기초지자체 방문자 수', '기초지자체 방문자 비율'], axis=1, inplace=True)
visit_df
```

```python
visit_df.rename(columns={'광역지자체명':'state', '기초지자체명':'city', '광역지자체 방문자 수':'visit_cnt'}, inplace=True)
visit_df
```

```python
visit_df = visit_df.astype({'visit_cnt':'int'})
```

```python
visit_df.dtypes
```

```python
# 고양시 테스트
goyang_df = visit_df[visit_df.city.str.contains('고양시')]
goyang_df
```

```python
goyang_sum = goyang_df.visit_cnt.sum()
```

```python
goyang_sum
```

```python
# '**구' 없애기
def remove_goo(city):
    if city.endswith('구'):
        city_split = city.split()
        no_goo = city_split[0]
        return no_goo
    else:
        return city
```

```python
visit_df.city = visit_df.city.apply(lambda x: remove_goo(x))
```

```python
visit_df
```

```python
# 잘 들어갔는지 확인
visit_df.to_csv('check1.csv')
```
